```python
# Interpolation Search and Binary Search Comparison
# Student Roll Numbers Range: 1 to 10000

def interpolation_search(arr, key):
    low = 0
    high = len(arr) - 1
    probes = 0

    while low <= high and key >= arr[low] and key <= arr[high]:
        probes += 1

        # Avoid division by zero
        if arr[high] == arr[low]:
            if arr[low] == key:
                return low, probes
            return -1, probes

        # Interpolation formula
        pos = low + int(
            ((high - low) / (arr[high] - arr[low])) * (key - arr[low])
        )

        if arr[pos] == key:
            return pos, probes

        elif arr[pos] < key:
            low = pos + 1

        else:
            high = pos - 1

    return -1, probes


def binary_search(arr, key):
    low = 0
    high = len(arr) - 1
    probes = 0

    while low <= high:
        probes += 1
        mid = (low + high) // 2

        if arr[mid] == key:
            return mid, probes

        elif arr[mid] < key:
            low = mid + 1

        else:
            high = mid - 1

    return -1, probes


# Example sorted roll numbers
roll_numbers = list(range(1, 10001))  # 1 to 10000

# Input roll number to search
key = int(input("Enter roll number to search: "))

# Interpolation Search
i_index, i_probes = interpolation_search(roll_numbers, key)

# Binary Search
b_index, b_probes = binary_search(roll_numbers, key)

# Results
print("\n--- Search Results ---")

if i_index != -1:
    print(f"Interpolation Search: Roll number found at index {i_index}")
else:
    print("Interpolation Search: Roll number not found")

print(f"Interpolation Search Probes: {i_probes}")

if b_index != -1:
    print(f"Binary Search: Roll number found at index {b_index}")
else:
    print("Binary Search: Roll number not found")

print(f"Binary Search Probes: {b_probes}")

# Comparison
print("\n--- Comparison ---")
if i_probes < b_probes:
    print("Interpolation Search required fewer probes.")
elif i_probes > b_probes:
    print("Binary Search required fewer probes.")
else:
    print("Both searches required the same number of probes.")
```
